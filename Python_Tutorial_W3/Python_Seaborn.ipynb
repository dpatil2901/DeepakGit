### Sample Datasets

    Seaborn provides several built-in datasets that we can use for data visualization and statistical analysis. 
These datasets are stored in pandas dataframes, making them easy to use with Seaborn's plotting functions.

Method used in seaborn to get  and load datasets:

    1. get_dataset_names()  : Method to display all available datasets.
    2. load_dataset()       : Method to load the specified dataset.This method takes the name of the dataset as argument.

##### get_dataset_names()

```python
# get_dataset_names() to display all the built-in datasets in Seaborn.

print(sns.get_dataset_names())
```

##### load_dataset()

### Seaborn Plot types

```python
Seaborn provides a wide range of plot types that can be used for data visualization and analysis. Seaborn plots are
broadly classified into 6 types.

 1. Relational plots
 2. Distribution Plots
 3. Categorical plots
 4. Regression plots
 5. Matrix plots
 6. Multi-plot grids

They are again sudivided into different plots as given below:
    
1. Relational plots

      1.  Scatter Plot
      2.  Line Plot
      3.  rel plot
    
2. Distribution Plots  

      1. displot
      2. histplot
      3. kdeplot (Kernel Density Estimation)
      4. ecdfplot
      5. rugplot

3. Categorical plots

      1. catplot
      2. stripplot
      3. swarmplot
      4. boxplot
      5. violinplot
      6. boxenplot
      7. pointplot
      8. barplot
      9. countplot
    
 4. Regression plots

      1. lmplot
      2. regplot
      3. residplot
    
 5. Matrix plots

      1. heatmap
      2. clustermap
        
6. Multi-plot grids

      1. Facet grids
      2. Pair grids (pairplot , pairGrid)
      3. Joint grids(jointplot, jointGrid)

In this tutorial we will familiarize each plot with our built-in datasets.
```

```python
crash_mx = crash_df.corr()
crash_mx

#plt.figure(figsize=(8,6))
#sns.set_context('paper', font_scale=1.4)
sns.heatmap(crash_mx, annot=True, cmap='Blues')
plt.show()
```

```python
flights = sns.load_dataset('flights')
#flights

flights = flights.pivot_table(index = 'month', columns= 'year', values= 'passengers')


#sns.heatmap(flights, cmap = 'Blues')
sns.heatmap(flights, cmap = 'Blues',linecolor='White', linewidth=1)
plt.show()
```

```python
sns.clustermap(flights,cmap='Blues',standard_scale=1)
plt.show()
```

```python
# Form a facetgrid using columns with a hue
#tips_fg=sns.FacetGrid(tips_df,col='time', row= 'smoker')
tips_fg=sns.FacetGrid(tips_df,col='time', hue='smoker',height=5, aspect=1.5)

# map the above form facetgrid with some attributes
#tips_fg.map(plt.hist,'total_bill',bins=8)
tips_fg.map(plt.scatter,'total_bill','tip')
plt.show()
```
